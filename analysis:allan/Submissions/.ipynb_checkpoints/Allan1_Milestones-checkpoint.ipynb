### Research Questions:

#### Question 1: Is there evidence to suggest that ethnic minorities are disproportionately affected by fatal encounters with law enforcement?


```python
import pandas as pd
import matplotlib.pyplot as plt
from Scripts import project_functions 
df = project_functions.load_and_process('fatal-police-shootings-data.csv')
fig, ethn=plt.subplots(figsize=(9,15))
explode = [0, 0.1, 0, 0, 0, 0]
colors = ["forestgreen", "turquoise", "orange", "pink", "magenta", "purple" ]
plt.title('Proportions of Fatal Police Encounters by Ethnicity')
df['ethnicity'].value_counts().plot(kind='pie', explode= explode, colors= colors, autopct='%1.0f%%', shadow=True)
df.ethnicity.value_counts(normalize=True)
```

#### Observation: Pie Chart
* From this pie chart we can see the share of police shootings by ethnicity. This visualization contributes to answering the question of how much more likely minority ethnicities are likely to have fatal police encounters. Caucasian ethnicities marked by the W make up a majority of cases with 51% of cases happening within this demographic. However, it is alarming that African Americans and Hispanic ethnicities make up 26% and 19% of fatal police shootings. Considering they are minority groups in the US it is more apparent that they make up a disproportional amount of fatal shooting cases. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from Scripts import project_functions 
df = project_functions.load_and_process('fatal-police-shootings-data.csv')

df['Date'] = pd.to_datetime(df['date'])
df.Date.dt.date

sns.set_theme(style="ticks", font_scale=1,)
cov = sns.displot(data = df, x="Date", hue="ethnicity", multiple="dodge")
plt.title('Number of Police Shooting Fatalities Per Year')
cov.set(xlabel= 'Years of Cases', ylabel='Fatal Police Shooting')
plt.show()
```

#### Observation: Countplot in years by Ethnicity.
* The number of police shootings seems fairly constant through the past 5 years. There are notable differences in 2020 most probably due to the coronavirus pandemic. Likewise, the number of victims in each ethnicity does not make any significant leaps aside from 2019 where African American fatalities reached a new high. On any given year Caucasian demographic had the greatest number of victims.

#### Question 2: How do the cases of fatal police shootings differ affect ethnicities in each state state?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from Scripts import project_functions 
df = project_functions.load_and_process('fatal-police-shootings-data.csv')




cov = plt.subplots(figsize=(20, 14))
sns.set_theme(style="white", font_scale=1)

cov = sns.countplot(data = df, x="state", hue="ethnicity")

plt.title('Number of Police Shooting Fatalities by State and Ethnicity')
cov.set(xlabel= 'State', ylabel='Fatal Police Shooting')

plt.show()
```

#### Observation: Countplot of Victims per State by Ethnicity.
* This visualization highlights the varying ethnic composition of victims per state. A notable observation as pointed out previously are the high number of cases in California(CA) where Hispanic ethnicities make up a significant majority of victims. This is also true for New Mexico however the number of cases in there are much lower than California. Additionally in Illinois(IL) and Louisiana (LA) African American ethnicities are the majority victims of fatal police encounters. However, it is important to note that African American cases are frequently high in many states aside from California and Louisiana.
